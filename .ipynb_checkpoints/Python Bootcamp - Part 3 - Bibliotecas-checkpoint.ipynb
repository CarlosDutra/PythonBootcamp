## Matplotlib

Biblioteca gráfica de Python amplamente utilizada. Auxilia nas análises de dados e geração de relatórios.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

A `Matplotlib` foi uma das bibliotecas pioneiras em realizar visualizações de forma bastante flexível, é possível fazer praticamente qualquer visualização nela, porém isso nem sempre é tão simples, conhecer todas as funcionalidades dela pode ser bastante desafiador.

Exibiremos agora alguns exemplos.

```python
x = np.linspace(-10,10,num = 21)
y = x**2

plt.plot(x,y)
plt.ylabel('y = x*x')
plt.show()
```

```python
z = 3*x**3 + 2*x**2 + 3*x + 2

plt.plot(x,z)
plt.ylabel('z = 3*x**3 + 2*x**2 + 3*x + 2')
plt.show()
```

É possível personalizar os gráficos utilizando os diversos parâmetros da biblioteca, como abaixo.

```python
plt.plot(x,z, 'ro')
plt.ylabel('z = 3*x**3 + 2*x**2 + 3*x + 2')
plt.show()
```

```python
plt.bar(x, z, align='center', alpha=0.5)
plt.show()
```

```python
plt.barh(x, z, align='center', alpha=0.5)
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal') 
plt.show()
```

Mais detalhes sobre a documentação da biblioteca podem ser encontrados no seguinte site: https://matplotlib.org/users/pyplot_tutorial.html.

## Seaborn

De forma complementar, uma outra biblioteca muito utilizada é a *Seaborn*.

```python
import seaborn as sns
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True);
```

## Vetorização

Para analisar quais clientes realizaram um agendamento de viagem, podemos criar uma flag auxiliar, da seguinte maneira.

```python
strftime("%Y-%m-%d %H:%M:%S", gmtime())
n = users.shape[0]
users['flag_booking'] = np.zeros(n)
for i in range(n):
    if users.loc[i,'country_destination'] == "NDF":
        users.loc[i,'flag_booking'] = 0
    else:
        users.loc[i,'flag_booking'] = 1
strftime("%Y-%m-%d %H:%M:%S", gmtime())
```

```python
users['flag_booking'].describe()
```

Uma forma alternativa muito mais eficiente é através de vetorização.

```python
users['flag_booking2'] = np.where(users['country_destination'] == "NDF", 0, 1)
```

```python
users['flag_booking2'].describe()
```

## Exercício

* Crie uma flag para os clientes que viajaram para os Estados Unidos
* Crie uma visualização para cálculo do volume que possui e não possui a flag.
* Crie um cubo para análise do volume que viajou para os Estados Unidos por `gênero`.

```python
users['flag_US'] = np.where(users['country_destination'] == "US", 0, 1)
```

```python
users.flag_US.value_counts(dropna=False).plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.xlabel('flag_US')
```

```python
cubo = users.groupby(['flag_US']).aggregate({'age': ['count']})
```

```python
cubo
```

```python
users.gender.value_counts(dropna=False).plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.xlabel('gender')
```

```python
cubo = users.groupby(['flag_US', 'gender']).aggregate({'age': ['count']})
```

```python
cubo
```
