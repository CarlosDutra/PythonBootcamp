## Matplotlib

Biblioteca gráfica de Python amplamente utilizada. Auxilia nas análises de dados e geração de relatórios.

```python
import matplotlib.pyplot as plt
```

A `Matplotlib` foi uma das bibliotecas pioneiras em realizar visualizações de forma bastante flexível, é possível fazer praticamente qualquer visualização nela, porém isso nem sempre é tão simples, conhecer todas as funcionalidades dela pode ser bastante desafiador.

Exibiremos agora alguns exemplos.

```python
x = np.linspace(-10,10,num = 21)
y = x**2

plt.plot(x,y)
plt.ylabel('y = x*x')
plt.show()
```

```python
z = 3*x**3 + 2*x**2 + 3*x + 2

plt.plot(x,z)
plt.ylabel('z = 3*x**3 + 2*x**2 + 3*x + 2')
plt.show()
```

É possível personalizar os gráficos utilizando os diversos parâmetros da biblioteca, como abaixo.

```python
plt.plot(x,z, 'ro')
plt.ylabel('z = 3*x**3 + 2*x**2 + 3*x + 2')
plt.show()
```

```python
plt.bar(x, z, align='center', alpha=0.5)
plt.show()
```

```python
plt.barh(x, z, align='center', alpha=0.5)
plt.show()
```

Mais detalhes sobre a documentação da biblioteca podem ser encontrados no seguinte site: https://matplotlib.org/users/pyplot_tutorial.html.

```python
import pandas as pd
```

### Agrupamento

Em Python é possível realizar operações de agrupamento.
